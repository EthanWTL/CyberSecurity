```python
data.columns = data.columns.str.replace(' ', '')
```

```python
df = data.drop(columns = ['FlowID', 'SourceIP', 'SourcePort','DestinationIP', 'DestinationPort', 'Protocol', 'Timestamp'])
```

```python
df.shape
```

```python
df['Label'].value_counts()
```

```python
df = df.copy()
cat_columns = df.select_dtypes(['object']).columns
cat_columns
```

```python
df[cat_columns] = df[cat_columns].astype('category')
df[cat_columns] = df[cat_columns].apply(lambda x: x.cat.codes)
df
data = df
data.dtypes
```

```python
df.head()
```

```python
df.tail()
```

```python
data_real = df.replace(np.inf, np.nan)
data_real.isnull().sum().sum()
```

```python
data_df = data_real.dropna(axis=0)
data_df.isnull().sum().sum()
```

```python
df = data_df
```

```python
df = df.sample(n=len(df), random_state=2)
```

```python
df.head()
```

```python
df['Label'].value_counts()
```

```python
X = df.iloc[:, 0:-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
X= standardScaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.67, test_size = 0.33, random_state = 42)
```

```python
X_train.shape
```

```python
X_train.reshape(463242, 77, 1)
```

```python
#CNN algorithm libraries
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Conv1D, GlobalMaxPooling1D,MaxPool1D,Flatten


#for evaluation of trained algorithms 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
from tensorflow.keras.optimizers import RMSprop
```

```python
model = Sequential()

model.add(Conv1D(filters=32, kernel_size=5, activation='relu', input_shape= (77,1)))
model.add(Conv1D(filters=32, kernel_size=5, activation='relu'))
model.add(Dropout(0.5))
model.add(MaxPool1D(pool_size=2))
model.add(Flatten())
model.add(Dense(30,activation='relu'))
model.add(Dense(6,activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
import time

start = time.time()
print("Timer Begin")

model.fit(X_train,y_train,batch_size=20,epochs=20,
          validation_split=0.1)

end = time.time()
print(end - start)
```

```python
model.save('multiclass_Dos.h5')
```
